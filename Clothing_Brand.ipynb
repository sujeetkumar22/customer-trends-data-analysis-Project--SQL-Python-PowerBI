```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))    # # now we are replacing the null values i the rating section with the mrdion of  their product category
```

```python
df.isnull().sum()
```

```python
# Best to use snake caseing so that the format becomes consistent and we get a
# formating that is easy to read
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('purchase_amount_(usd)' , 'purchase_amount', regex=False)
```

```python
df.columns
```

```python
#Create a column age_group
labels = ['Young Adult', 'Adult' , 'Middle' , 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create column purchase_frequency_days only for the making our work easier
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly': 7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually': 365,
    'Every 3 Monthly':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df=df.drop('promo_code_used',axis=1, errors='ignore')
df.columns
```
